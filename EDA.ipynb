```python
#import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
df = pd.read_csv (r'C:\Users\Tombra\Water-potability-prediction-\water potability prediction.csv')
```

```python
# display the dataframe
df
```

```python
# Create a new column 'normalized_conductivity' by dividing 'conductivity' by 100
df['Normalized_Conductivity (µS/m)'] = df['Conductivity (µS/cm)'] / 100
```

```python
# perform statistical analysis 
df.describe()
```

```python
# check for null values
df.isnull().sum()
```

```python
# Histograms for each column
# Select the columns for EDA, excluding 'Sample ID'
columns_for_eda = ['Dissolved Oxygen (mg/L)', 'pH', 'Temperature (°C)', 'Turbidity (NTU)', 'Normalized_Conductivity (µS/m)']

# Generate histograms for the selected columns
df[columns_for_eda].hist(figsize=(10, 6))
plt.tight_layout()
plt.show()


# Boxplot for each column
df[columns_for_eda].boxplot(figsize=(10, 6))
plt.tight_layout()
plt.show()

# Correlation heatmap
corr_matrix = df[columns_for_eda].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Identify outliers using boxplots in the "dissolved oxygen" column
sns.boxplot(data=df, y="Dissolved Oxygen (mg/L)")
plt.title("Outliers in Dissolved Oxygen")
plt.show()
```

```python
# Plotting pH vs conductivity
plt.scatter(df['pH'], df['Normalized_Conductivity (µS/m)'])
plt.xlabel('pH')
plt.ylabel('Normalized_Conductivity (µS/m)')
plt.title('Relationship between pH and Conductivity')
plt.show()
```

```python
# Plotting pH vs conductivity
plt.scatter(df['pH'], df['Temperature (°C)'])
plt.xlabel('pH')
plt.ylabel('Temperature (°C)')
plt.title('Relationship between pH and Temperature')
plt.show()
```

```python
sns.regplot(x='pH', y='Normalized_Conductivity (µS/m)', data=df)
plt.xlabel('pH')
plt.ylabel('Normalized_Conductivity (µS/m)')
plt.title('Relationship between pH and Conductivity')
plt.show()
```
