```python
#import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
df = pd.read_csv (r'C:\Users\Tombra\Water-potability-prediction-\water_potability.csv')
```

```python
# Calculate the column averages
column_averages = df.mean()

# Replace null values with column averages
df1 = df.fillna(column_averages)


```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1['Normalized_solids'] = df1['Solids']/100
```

```python
df1.describe()
```

```python
# Histograms for each column
# Select the columns for EDA, excluding 'Sample ID'
columns_for_eda = ['ph', 'Hardness', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon','Trihalomethanes','Turbidity','Potability','Normalized_solids']

# Generate histograms for the selected columns
df1[columns_for_eda].hist(figsize=(10, 6))
plt.tight_layout()
plt.show()


# Boxplot for each column
df1[columns_for_eda].boxplot(figsize=(10, 6))
plt.tight_layout()
plt.show()

# Correlation heatmap
corr_matrix = df1[columns_for_eda].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
